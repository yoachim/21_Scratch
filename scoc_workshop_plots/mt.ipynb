```python
basedir = 'git_repos/survey_strategy/fbs_2.0'
runs = ri.FamilyInfo('runs_v2.0.json','summary_11_8.csv')
```

```python
baseline_df = runs.family_data('baseline')
```

```python
def trend_plots(baseline, runs, colnames=None, linewidth=5, alpha=0.8):
    
    if colnames == None:
        colnames = {'# SNe Ia': 'Total detected SNNSNMetric_nSN  HealpixSlicer',
                    'Microlensing Frac': 'Fraction detected of total (mean) 1 to 10 day Microlensing  UserPointsSlicer',
                   'Proper Motion inv': 'Median Proper Motion Error @ 20.5 All visits HealpixSlicer'}
    fig, ax = plt.subplots()
    for key in colnames:
        if 'Proper Motion' in key:
            ax.plot(baseline[colnames[key]].max()/runs[colnames[key]], 'o-', label=key,
                    linewidth=linewidth, alpha=alpha)
        else:
            ax.plot(runs[colnames[key]]/baseline[colnames[key]].max(), 'o-',label=key,
                    linewidth=linewidth, alpha=alpha)
        
    ax.legend()
    ax.axhline(1, linestyle='--', color='black')
    
    # Shorten up the labels a bit
    trim_ticks = [label.replace('_v2.0_10yrs','') for label in runs.index.values]
    
    ax.set_xticklabels(trim_ticks, rotation = 90)
    ax.set_ylabel('Metric Value / Baseline')
    return fig
```

```python
fig = trend_plots(baseline_df, runs.family_data('vary_nes'))
```

```python
fig = trend_plots(baseline_df, runs.family_data('vary_gp'))
```
