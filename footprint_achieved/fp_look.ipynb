```python
from rubin_sim.scheduler.utils import standard_goals
import numpy as np
import healpy as hp
import matplotlib.pylab as plt
%matplotlib inline

import rubin_sim.maf.db as db
import rubin_sim.maf.utils as utils
import rubin_sim.maf.metrics as metrics
import rubin_sim.maf.slicers as slicers
import rubin_sim.maf.stackers as stackers
import rubin_sim.maf.metricBundles as metricBundles
```

```python
sg = standard_goals()
```

```python
hp.mollview(sg['r'])
```

```python
dbFile = 'baseline_nexp2_v1.7.1_10yrs.db'
conn = db.OpsimDatabase(dbFile)
outDir='temp'
resultsDb = db.ResultsDb(outDir=outDir)
```

```python
bundleList = []
for filtername in 'ugrizy':
    sql = 'filter="%s" and note not like "%%DD%%"' % filtername
    metric = metrics.CountMetric(col='night')
    slicer = slicers.HealpixSlicer(nside=32)
    summaryStats = []
    plotDict = {}
    bundleList.append(metricBundles.MetricBundle(metric,slicer,sql, plotDict=plotDict, summaryMetrics=summaryStats))
```

```python


bd = metricBundles.makeBundlesDictFromList(bundleList)
bg = metricBundles.MetricBundleGroup(bd, conn, outDir=outDir, resultsDb=resultsDb)
bg.runAll()

bg.plotAll(closefigs=False)
```

```python
all_sum = bundleList[0].metricValues.data*0
for bun in bundleList:
    all_sum += bun.metricValues.filled(0)
```

```python
hp.mollview(all_sum, min=200, max=500)
```

```python
ack = plt.hist(all_sum, bins=100)
```
